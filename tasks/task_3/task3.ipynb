```python
import zipfile
from pathlib import Path
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') 
```

```python
zip_path = Path('../datasets/video_split.zip')
extract_path = Path('../datasets/video_split')
```

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
jpg_files = sorted(extract_path.rglob('*.jpg'))

for img_path in jpg_files:
    results = model(img_path, conf=0.4, save=True)
```

```python
import cv2
```

```python
frames_dir = Path("C:\\Users\\user1\\runs\\detect\\predict4")
output_path = Path("..\\tasks\\task_3\\results\\detection\\output_detected_video.mp4")
```

```python
frames = sorted(frames_dir.glob("*.jpg"))
first = cv2.imread(str(frames[0]))
h, w, _ = first.shape

out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), 30, (w, h))

for f in frames:
    frame = cv2.imread(str(f))
    out.write(frame)

out.release()
print(f"Annotated video saved at {output_path}")
```
